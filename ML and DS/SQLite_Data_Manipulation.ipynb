## Read Database and find dates with temperature difference of 10 degrees.

```python
import pandas as pd
import sqlite3
```

```python
df = pd.read_csv("weatherHistory.csv")
```

```python
df
```

```python
df['Temperature (F)'] = df['Temperature (C)'] * 1.8 + 32
```

```python
df
```

```python
df1 = df[["Formatted Date", "Temperature (F)"]]
```

```python
df1
```

```python
conn = sqlite3.connect("weather.db")
```

```python
df1.to_sql('info', conn,if_exists='replace')
```

```python
cur = conn.execute("select * from info;")
```

```python
cur.fetchall()
```

```python
cur.execute("PRAGMA table_info('info')").fetchall()
```

```python
diff_10 = cur.execute("""with myItems as (select *, "Temperature (F)" - lag("Temperature (F)", 1)
               over(order by "index") as val2 from info)
               select * from myItems WHERE val2 > 10""").fetchall()
diff_10
```

```python
all_row = cur.execute("select * from info ").fetchall()

        
```

```python
for i in all_row:
    if i[0] == 720 or i[0] == 721 or i[0] == 719:
        print(i)
```

```python
import csv
diff_10.insert(0, ('index', 'Date', 'Temperature(f)', "Difference"))
with open("only_10(f)_diff.csv", "w", newline="") as f:
    writer = csv.writer(f)
    writer.writerows(diff_10)
```

```python
conn.close()
```
